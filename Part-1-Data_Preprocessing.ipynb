```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv ("train.csv")
data_processed = data
data.head (3)
```

### Convert the MMM-YY, Date Of Joining and Last Working Date Columns to Datetime

```python
data ['MMM-YY'] = pd.to_datetime (data ['MMM-YY'])
data.Dateofjoining = pd.to_datetime (data.Dateofjoining)
data.LastWorkingDate = pd.to_datetime (data.LastWorkingDate)
data.head (5)
```

### Add column for No. of months the person worked

```python
emp_duration = ((data ['MMM-YY'] - data.Dateofjoining)/np.timedelta64 (1, 'M')).values
# There were certain dates where this will output a negative value, so to get rid of those :-
emp_duration = np.clip (emp_duration, 0, 100000)
emp_duration
```

```python
new_data = pd.DataFrame(emp_duration, columns = ['emp_duration'])
data_processed = pd.concat([data_processed, new_data], join = 'outer', axis = 1)
```

```python
data_processed.head(3)
```

### Find the Amount by which the salary of a person changed compared to the last month

```python
data_processed ['emp_salary_change'] = data.groupby ('Emp_ID')['Salary'].diff ()
```

```python
data_processed.head(3)
```

### Find the Amount by which the Rating of a person changed compared to the last month

```python
data_processed ['emp_rating_change'] = data.groupby ('Emp_ID')['Quarterly Rating'].diff ()
```

```python
data_processed.head(3)
```

### Find the Amount by which the Designation of a person changed compared to the last month

```python
data_processed ['emp_designation_change'] = data.groupby ('Emp_ID')['Designation'].diff ()
```

```python
data_processed.head (3)
```

### Find the Amount by which the Business Value of a person changed compared to the last month

```python
data_processed ['emp_business_value_change'] = data.groupby ('Emp_ID')['Total Business Value'].diff ()
```

```python
data_processed.head (3)
```

### Convert MMM-YY to Quarters

```python
data_processed ['emp_quarter_current'] = pd.PeriodIndex (data_processed ['MMM-YY'], freq = 'Q')
data_processed.head (3)
```

### Find the last working quarter of the employee

```python
x = data_processed.groupby ('Emp_ID')['LastWorkingDate'].max ()
x = pd.DataFrame (x)
x.head ()
```

#### For currently working employees (**NaT values**), set the last working day beyond the next 2 quarters

```python
end = pd.to_datetime ('2018-12-1')
x.LastWorkingDate = x.LastWorkingDate.fillna (end)
```

```python
x ['emp_quarter_last'] = pd.PeriodIndex (x.LastWorkingDate, freq = 'Q')
x = x.drop (columns = ['LastWorkingDate'])
x.head ()
```

```python
data_processed = pd.merge (data_processed, x, on = 'Emp_ID')
```

```python
data_processed.head (3)
```

### Find the Attrition value
##### Attrition = 1, if the person DOES NOT leave the company in the next 2 Quarters
##### Attrition = 0, if the person DOES leave the company in the next 2 Quarters

```python
attrition = np.where (data_processed ['emp_duration'] >= 6, 1, 0)
```

```python
attrition = []
for i in range (len (data_processed)) :
    day_0 = data_processed.emp_quarter_current[i].to_timestamp()
    day_1 = data_processed.emp_quarter_last[i].to_timestamp()
    months = (day_1 - day_0) / np.timedelta64(1, 'M')
    if months > 5.5 :
        attrition.append (1)
    else :
        attrition.append (0)
```

```python
data_processed ['attrition'] = attrition
```

```python
data_processed.head ()
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
data_processed.Gender = label_encoder.fit_transform (data_processed.Gender)
data_processed.City = label_encoder.fit_transform (data_processed.City)
data_processed.Education_Level = label_encoder.fit_transform (data_processed.Education_Level)
```
